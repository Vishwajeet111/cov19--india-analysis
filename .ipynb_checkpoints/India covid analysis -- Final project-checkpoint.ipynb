```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
cov_df = pd.read_csv('covid_19_india.csv')
cov_df.head()
```

```python
cov_df.shape
```

```python
cov_df.columns
```

```python
cov_df.info()
```

```python
cov_df.isnull().sum()
```

```python
cov_df['State/UnionTerritory'].unique()
```

```python
# Replacing all the improper state name with proper name
cov_df['State/UnionTerritory'].replace(['Bihar****','Telengana'],['Bihar','Telangana'],inplace=True)
```

```python
cov_df['State/UnionTerritory'].replace(['Daman & Diu'],['Dadra and Nagar Haveli and Daman and Diu'],inplace=True)
```

```python
cov_df['State/UnionTerritory'].replace(['Dadra and Nagar Haveli'],['Dadra and Nagar Haveli and Daman and Diu'],inplace=True)
```

```python
# Removing all the invalid values from State/UnionTerritory column
unassigned_index = cov_df[cov_df['State/UnionTerritory'] == 'Unassigned'].index
reassigned_index = cov_df[cov_df['State/UnionTerritory'] == 'Cases being reassigned to states'].index
cov_df['State/UnionTerritory'] = cov_df['State/UnionTerritory'].drop(unassigned_index)
cov_df['State/UnionTerritory'] = cov_df['State/UnionTerritory'].drop(reassigned_index)
```

```python
cov_df['State/UnionTerritory'].value_counts()
```

```python
cov_df['Time'].unique()
```

```python
# removing all the useless columns -- Sno, Time, ConfirmedIndianNational, ConfirmedForeignNational
cov_df.drop(['Sno','Time','ConfirmedIndianNational','ConfirmedForeignNational'],axis=1,inplace=True)
```

```python
cov_df.head()
```

```python
cov_df['ActiveCases'] = cov_df['Confirmed'] - (cov_df['Cured'] + cov_df['Deaths'])
```

```python
cov_df.head()
```

```python
cov_df.describe()
```

```python
state_sum_df = cov_df.groupby('State/UnionTerritory').sum().sort_values('Cured').reset_index()
```

```python
plt.rcParams['figure.figsize'] = (20,12)
for i,j in enumerate(state_sum_df.columns[1:]):
    plt.subplot(2,2,i+1)
    plt.pie(x=state_sum_df[j][20:],labels=state_sum_df['State/UnionTerritory'][20:],autopct='%.2f%%')
    plt.title('Percentage of {} cases'.format(j))
```

* From the above pie chart, we can observe that percentage of death cases in Maharashtra is more compared to other states.
* Also the percentage of cured cases is less compared to death cases in Maharashtra.
* After Maharashtra, Karnataka Tamilnadu Kerala and Andhra Pradesh have more ActiveCases.
* For Kerala we can observe that percentage of death cases is less compared to Active cases and cured cases which is good.
* For Andhra Pradesh, we can observe that percentage of cured cases is more compared to active cases and death cases which is very good
* From the overall pie chart, we can say that Maharashtra is most affected state followed by Karnataka and Tamilnadu.

```python
cov_df['Date'] = pd.to_datetime(cov_df['Date'])
cov_df['Year'] = cov_df['Date'].dt.year
cov_df['Month'] = cov_df['Date'].dt.month
cov_df['Month-Year'] = cov_df['Date'].dt.to_period('M')
cov_df.head()
```

```python
sns.countplot(x='Year',data=cov_df)
plt.xlabel('Year')
plt.ylabel('Count of rows')
plt.title('Distribution of data wrt Year')
plt.show()
```

```python
month_year_df = cov_df.groupby('Month-Year').sum().reset_index()
month_year_df['Month-Year'] = month_year_df['Month-Year'].astype(str)
month_year_df
```

```python
plt.subplot(3,1,1)
sns.lineplot(x='Date',y='ActiveCases',color='red',data=cov_df)
plt.title('Timeline of daily increase of Active Cases in India from 2020 to 2021')
plt.tight_layout()
plt.subplot(3,1,2)
sns.lineplot(x='Date',y='Cured',color='green',data=cov_df)
plt.title('Timeline of daily increase of Cured cases in India from 2020 to 2021')
plt.tight_layout()
plt.subplot(3,1,3)
sns.lineplot(x='Date',y='Deaths',data=cov_df)
plt.title('Timeline of daily increase of Death cases in India from 2020 to 2021')
```

* From the above timeline we can observe that no. of active cases are slowly increasing from July 2020 which marks the starting of first wave. After December 2020 the active cases started decreasing slowly.
* Between July 2020 and December 2020 the number of cured case also started increasing which is a good but during this the number of death cases also increased which is bad.
* Again from April 2021 the number of active cases started to increase and maybe at June 2021 the number of active cases reached at highest peak and then the cases started to decrease slowly.
* The number of cured cases increased at 2021 compared to 2020.

```python
plt.subplot(3,1,1)
sns.lineplot(x='Month-Year',y='ActiveCases',color='red',data=month_year_df)
plt.title('Timeline of monthly increase of Active Cases in India from 2020 to 2021')
plt.tight_layout()

plt.subplot(3,1,2)
sns.lineplot(x='Month-Year',y='Cured',color='green',data=month_year_df)
plt.title('Timeline of monthly increase of Cured cases in India from 2020 to 2021')
plt.tight_layout()

plt.subplot(3,1,3)
sns.lineplot(x='Month-Year',y='Deaths',data=month_year_df)
plt.title('Timeline of monthly increase of Death cases in India from 2020 to 2021')
```

* Same observation done above

```python
cured_20 = cov_df[cov_df['Year'] == 2020].Cured.sum()
death_20 = cov_df[cov_df['Year'] == 2020].Deaths.sum()
active_20 = cov_df[cov_df['Year'] == 2020].ActiveCases.sum()

cured_21 = cov_df[cov_df['Year'] == 2021].Cured.sum()
death_21 = cov_df[cov_df['Year'] == 2021].Deaths.sum()
active_21 = cov_df[cov_df['Year'] == 2021].ActiveCases.sum()

plt.subplot(1,2,1)
x1 = [cured_20,death_20,active_20]
labels = ['Cured Cases','Death Cases','Active Cases']
plt.pie(x=x1,labels=labels,shadow=True,autopct='%.2f%%')
plt.title('Percentage distribution of cases in India for 2020')

plt.subplot(1,2,2)
x2 = [cured_21,death_21,active_21]
labels = ['Cured Cases','Death Cases','Active Cases']
plt.pie(x=x2,labels=labels,shadow=True,autopct='%.2f%%')
plt.title('Percentage distribution of cases in India for 2021')
plt.show()
```

* From the above piechart we can observe that the percentage of the active cases decreased from year 2020 to 2021 and percentage of cured cases increased.
* Also the percentage of death cases decreased in year 2021 which is a very good sign.
* Overall the condition of India was better in 2021 compared to 2020.

* From the above piechart we can observe that the percentage of the active cases decreased from year 2020 to 2021 and percentage of cured cases increased.
* Also the percentage of death cases decreased in year 2021 which is a very good sign.
* Overall the condition of Maharashtra was better in 2021 compared to 2020.
