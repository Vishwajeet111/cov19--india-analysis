```python
import pandas as pd
import numpy as np
import seaborn as sns
import plotly as plt
import plotly.express as px
```

```python
df = pd.read_csv('covid_19_india.csv')
df.head()
```

```python
df.dtypes
```

```python
df.shape  #(row,column)
```

```python
df['State/UnionTerritory'].unique()
```

```python
Bihar**** --> Bihar
Telengana --> Telangana 
```

```python
df['State/UnionTerritory'].replace('Bihar****','Bihar',inplace=True)
df['State/UnionTerritory'].replace('Telengana','Telangana',inplace=True)
```

```python
df['State/UnionTerritory'].unique()
```

```python
# 'Daman and Nagar Haveli and Daman and Diu' <=== 'Dadra and Nagar Haveli' , 'Daman & Diu'
```

```python
df[df['State/UnionTerritory'] == 'Dadra and Nagar Haveli and Daman and Diu']
```

```python
df['State/UnionTerritory'].replace(['Dadra and Nagar Haveli','Daman & Diu'],['Dadra and Nagar Haveli and Daman and Diu','Dadra and Nagar Haveli and Daman and Diu'],inplace=True)
```

```python
df1 = df[df['State/UnionTerritory'] != 'Cases being reassigned to states']
```

```python
# df1['State/UnionTerritory'].unique()
```

```python
df1 = df1[df1['State/UnionTerritory'] != 'Unassigned']
```

```python
df1['State/UnionTerritory'].unique()
```

```python
df1[['Cured','Confirmed','Deaths']].isnull().sum()
```

```python
df1.describe()
```

```python
df1.groupby('State/UnionTerritory').Deaths.mean()
# df1.groupby('State/UnionTerritory').Confirmed.mean()
# df1.groupby('State/UnionTerritory').Cured.mean()
```

```python
# df1.groupby('State/UnionTerritory').Deaths.sum()
df1.groupby('State/UnionTerritory').Confirmed.sum()
# df1.groupby('State/UnionTerritory').Cured.mean()
```
